```python
import pandas as pd

# 1. Panggil data bersih yang kemarin udah kita save!
df = pd.read_csv('startup_cleaned.csv')

# 2. Ubah kolom 'founded_at' yang tadinya cuma teks biasa jadi format WAKTU (Datetime) beneran
df['founded_at'] = pd.to_datetime(df['founded_at'], errors='coerce')

# 3. MENCIPTAKAN FITUR BARU: Umur Startup!
# (Dataset Crunchbase ini direkam sekitar akhir tahun 2015, jadi kita pakai 2015 sebagai patokan umur)
df['umur_startup_tahun'] = 2015 - df['founded_at'].dt.year

# 4. Intip hasilnya!
df[['name', 'founded_at', 'umur_startup_tahun']].head(10)
```

```python
# 1. Buang startup yang umur atau total pendanaannya kosong (NaN)
df = df.dropna(subset=['umur_startup_tahun', 'funding_total_usd'])

# 2. Pilih "Tim Inti" (Fitur) yang mau kita ajarin ke AI
fitur_pilihan = ['funding_total_usd', 'funding_rounds', 'umur_startup_tahun', 'status']
df_final = df[fitur_pilihan].copy()

# 3. TRANSLASI BAHASA AI! Ubah target teks jadi angka
# Bangkrut (closed) = 0, Sukses (acquired) = 1, Masih Jalan (operating) = 2
status_map = {'closed': 0, 'acquired': 1, 'operating': 2}
df_final['target'] = df_final['status'].map(status_map)

# Buang kolom status lama yang masih bentuk teks
df_final = df_final.drop(columns=['status'])

# Intip bentuk akhir data kita sebelum dimasukin ke mesin AI
print("Jumlah data siap tempur:", df_final.shape)
df_final.head()
```

```python
# Save data yang udah jadi angka semua ini!
df_final.to_csv('startup_ml_ready.csv', index=False)

print("Data ML-Ready berhasil disave! AI siap dilatih! 🤖💾")
```
